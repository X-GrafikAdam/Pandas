## Fonksiyon Uygulamak

**Ülkelere Göre Bölünmüş Median maaş**

```python
country_group = df.groupby(["Country"]) #ülkelerin maaşların medyanlarıı 
```

```python
country_group["Salary"].median() #alttaki tablo bir seriesdir.
```

```python
country_group["Salary"].median().sort_values(ascending = False)
```

```python
country_group["Salary"].median()["United States"]
```

```python
median_country = country_group["Salary"].median()#yukarıdakileri yazmak yerine median country e eşitle. bu daha basit...
```

```python
median_country["Turkey"]
```

```python
median_country["Germany"]
```

### agg()

Birden çok fonksiyon uygulamak istiyorsak

```python
#agg() diye bir fonksiyon var.
#birden çok series dataframe oluşturur.
```

```python
country_group["Salary"].agg(["median", "mean"])
```

```python
country_group["Salary"].agg(["median", "mean"]).loc[["Turkey", "Germany"]]
```

```python
type(country_group["Salary"].agg(["median", "mean"]))
```

```python
country_group["Salary"].agg(["median", "mean"]).loc["Germany"]
```

```python
type(country_group["Salary"].agg(["median", "mean"]).loc["Germany"])
```

```python
hobby_group = df.groupby(["Hobbyist"])
```

```python
hobby_group["Salary"].agg(["median", "mean"])
```
